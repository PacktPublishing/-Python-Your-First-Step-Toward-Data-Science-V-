# Visualizing summary statistics on your datasets

## Visualizing data with third party packages

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 1000)
y = np.cumsum(np.random.randn(1000, 6), 0)
```

```python
plt.plot(x, y)
plt.show()
```

```python
import seaborn as sns
sns.set()
```

```python
plt.plot(x, y)
plt.show()
```

```python
data = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])

for dataset in 'ab':
    plt.hist(data[dataset], normed=True, alpha=0.5)
    
plt.show()
```

```python
for dataset in 'ab':
    sns.kdeplot(data[dataset], shade=True)    
plt.show()
```

```python
sns.distplot(data['a'])
```

```python
sns.jointplot("a", "b", data, kind='kde')
```

```python
sns.jointplot("a", "b", data, kind='hex')
```

# Exploring a dataset with Pandas

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
features, labels = load_iris(return_X_y=True)
```

```python
df = pd.DataFrame(features)
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.75, random_state=42)
```

```python
df = pd.DataFrame(X_train)
df["label"] = y_train
df
```

```python
df.head()
```

```python
df.mean(0)
```

```python
df.mean(1)
```

```python
df.describe()
```

```python
print(df.idxmax())
print(df.idxmin())
```

```python
df.loc[:, 2].value_counts()
```

```python
pd.cut(df.loc[:, 2],4)
```

```python
pd.cut(df.loc[:, 2],4).value_counts()
```




# Visualizing Summary Statistics

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
ts = pd.Series(np.random.randn(1000))
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=list('ABCD'))
df = df.cumsum()
df.plot();
```

```python
df.plot(x='A', y='B');
```

```python
df.iloc[15].plot(kind='bar');
```

```python
df = pd.DataFrame(np.random.randn(10, 4), columns=list('ABCD'))
df.plot.bar();
```

```python
df.plot.bar(stacked=True);
```

```python
df.plot.barh(stacked=True);
```

```python
df.plot.hist();
```

```python
df.plot.hist(stacked=True);
```

```python
df.plot.box()
```

```python
pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd']).plot.area();
```
